```python
import os
import numpy as np
```

```python
all_path = '../SLR_dataset/seq_txt'
#读取全部数据的根目录
```

```python
#获取全部数据的文件夹
folder_list = os.listdir(all_path)
```

```python
# 获取全部数据文件夹的完整目录
folder_list = [os.path.join(all_path,i) for i in folder_list]
```

```python
def left_right_hand(data):
    left = data[0:63]
    right = data [63:]
    return left,right
```

```python
def list_proLR(data):
    #该函数用于将读取到的数据变成左右手格式
    result = []
    for i in data:
        left,right = left_right_hand(i)
        result.append(left)
        result.append(right)
    return result
```

```python
all_data = []
for folder in folder_list:
    data_list = [os.path.join(folder,filename) for filename in os.listdir(folder) if filename.endswith('.avi')]
    for file in data_list:
        # print(file)
        data = []
        with open(file,'r') as f:
            for line in f:
                float_values = [float(x) for x in line.split()]
                data.append(float_values)
        # print(len(data)) #二维数组 帧*126
        # print(data[0:10])
        all_data += list_proLR(data) #三维数组 帧*2*63
```

```python
all_data=np.array(all_data)
```

```python
np.save('../trial_sheet/准确率正常的全部数据.npy', all_data)
```

```python
from sklearn.cluster import MiniBatchKMeans
import numpy as np

# 以内存映射方式加载 .npy 文件
data = np.load('../trial_sheet/准确率正常的全部数据.npy', mmap_mode='r')

# 初始化 MiniBatchKMeans
n_clusters = 20
batch_size = 10000  # 每批读取和处理的样本数
kmeans = MiniBatchKMeans(n_clusters=n_clusters, batch_size=batch_size, random_state=42)

# 分批读取数据并拟合模型
num_samples = data.shape[0]  # 总样本数
for i in range(0, num_samples, batch_size):
    batch = data[i:i + batch_size]  # 读取当前批次的数据
    kmeans.partial_fit(batch)  # 使用当前批次的数据更新模型

    # 每处理 30 个批次打印一次信息
    if (i // batch_size + 1) % 30 == 0:
        print(f"已处理 {i // batch_size + 1} 个批次，当前批次形状: {batch.shape}")

# 获取聚类结果
labels = kmeans.labels_  # 每个样本的聚类标签
centers = kmeans.cluster_centers_  # 聚类中心

# 输出结果
print("聚类标签:", labels)
print("聚类中心:", centers)
```

```python
from sklearn.cluster import MiniBatchKMeans
import numpy as np

# 以内存映射方式加载 .npy 文件
data = np.load('../trial_sheet/准确率正常的全部数据.npy', mmap_mode='r')

# 初始化 MiniBatchKMeans
n_clusters = 40
batch_size = 10000  # 每批读取和处理的样本数
kmeans = MiniBatchKMeans(n_clusters=n_clusters, batch_size=batch_size, random_state=42)

# 分批读取数据并拟合模型
num_samples = data.shape[0]  # 总样本数
for i in range(0, num_samples, batch_size):
    batch = data[i:i + batch_size]  # 读取当前批次的数据
    kmeans.partial_fit(batch)  # 使用当前批次的数据更新模型

    # 每处理 30 个批次打印一次信息
    if (i // batch_size + 1) % 30 == 0:
        print(f"已处理 {i // batch_size + 1} 个批次，当前批次形状: {batch.shape}")

# 获取聚类结果
labels = kmeans.labels_  # 每个样本的聚类标签
centers = kmeans.cluster_centers_  # 聚类中心

# 统计每个标签的点数
unique_labels, counts = np.unique(labels, return_counts=True)

# 打印每个标签的点数
print("\n每个聚类标签的点数统计:")
a = 0
for label, count in zip(unique_labels, counts):
    print(f"标签 {label}: {count} 个点")
    a+=count
print(a)
```
