```python
import matplotlib.pyplot as plt

# 示例2D列表数据，包含两组(x, y)数据
data = [
    #第一条线，emb128，hid128
    [(40, 0.657), 
     (50, 0.658),
     (70, 0.655),
     (100, 0.641)],
    #第二条线，emb256,hid256
    [(20, 0.650), (40, 0.668), (50, 0.661),(70,0.6556),(100,0.655)],
    #第三条线，emb512,hid512
    [(30,0.647),(40,0.669),(70,0.664),(100,0.662)],
    #第四条线 emb 512 hid 256
    [(50,0.660),],
]

# 绘制每条折线
for d in data:
    x_values = [point[0] for point in d] # 提取x值
    y_values = [point[1] for point in d] # 提取y值
    plt.plot(x_values, y_values, marker='o') # 绘制折线，并在点上显示小圆圈标记

# 添加标题和轴标签
plt.title('2D')
plt.xlabel('emb')
plt.ylabel('bleu')
# plt.ylim(0.5, 1)

# 显示图表
plt.legend(['emb128 hid128', 'emb256 hid256','emb512 hid512','emb512 hid256']) # 如果有多条线，可以添加图例
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 数据解析
data = [
    # 格式: [模型结构, 训练集占比, 模型参数, 双手合并, 抽帧, 参数量, 参数量(每秒), bleu]
    ['Lstm', 0.8, 'Km20-emb256-hid256', False, 1, 1151157, 55255536, 0.650],
    ['Lstm', 0.8, 'Km30-emb512-hid512', False, 1, 4404405, 211411440, 0.6454],
    ['Lstm', 0.8, 'Km40-emb128-hid128', False, 1, 316085, 15172080, 0.657],
    ['Lstm', 0.8, 'Km40-emb256-hid256', False, 1, 1156277, 55501296, 0.668],
    ['Lstm', 0.8, 'Km40-emb512-hid512', False, 1, 4409525, 211657200, 0.669],
    ['Lstm', 0.8, 'Km50-emb128-hid128', False, 1, 317365, 15233520, 0.658],
    ['Lstm', 0.8, 'Km50-emb256-hid256', False, 1, 1158837, 55624176, 0.661],
    ['Lstm', 0.8, 'Km70-emb128-hid128', False, 1, 319925, 15356400, 0.656],
    ['Lstm', 0.8, 'Km70-emb256-hid256', False, 1, 1163957, 55869936, 0.655],
    ['Lstm', 0.8, 'Km70-emb512-hid512', False, 1, 4424885, 212394480, 0.667],
    ['Lstm', 0.8, 'Km100-emb128-hid128', False, 1, 323765, 15540720, 0.641],
    ['Lstm', 0.8, 'Km100-emb256-hid256', False, 1, 1171637, 56238576, 0.655],
    ['Lstm', 0.8, 'Km100-emb512-hid512', False, 1, 4440245, 213131760, 0.662],
    ['Lstm', 0.8, 'Km80-emb512-hid512', True, 1, 4430005, 106320120, 0.626],
    ['Lstm', 0.8, 'Km80-emb512-hid512', True, 3, 4430005, 35440040, 0.632],
    ['Lstm', 0.8, 'Km80-emb512-hid512', True, 5, 4430005, 21264024, 0.627],
    ['Lstm', 0.8, 'Km80-emb512-hid512', True, 7, 4430005, 15188588, 0.617],
    
    ['Lstm', 0.8, 'kme80_emb512_hid512', True, 5, 4430005, 21264024, 0.6272932],
    ['Lstm', 0.5, 'kme80_emb512_hid512', True, 5, 4430005, 21264024, 0.6105132],
    ['Lstm', 0.3, 'kme80_emb512_hid512', True, 5, 4430005, 21264024, 0.5512863],
    ['Lstm', 0.2, 'kme80_emb512_hid512', True, 5, 4430005, 21264024, 0.4609284],
    ['Lstm', 0.8, 'kme80_emb512_hid512_multi-head', True, 5, 10469045, 50251416, 0.6309798],
    ['Lstm', 0.5, 'kme80_emb512_hid512_multi-head', True, 5, 10469045, 50251416, 0.618],
    ['Lstm', 0.3, 'kme80_emb512_hid512_multi-head', True, 5, 10469045, 50251416, 0.5831789],
    ['Lstm', 0.2, 'kme80_emb512_hid512_multi-head', True, 5, 10469045, 50251416, 0.5111360],
    ['Lstm', 0.8, 'kme40_emb256_hid256_multi-head', True, 1, 2668469, 64043256, 0.5959083],
    ['Lstm', 0.5, 'kme40_emb256_hid256_multi-head', True, 1, 2668469, 64043256, 0.595],
    ['Lstm', 0.3, 'kme40_emb256_hid256_multi-head', True, 1, 2668469, 64043256, 0.5611050],
    ['Lstm', 0.8, 'kme40_emb256_hid256', True, 1, 319925, 7678200, 0.595],
    ['Lstm', 0.5, 'kme40_emb256_hid256', True, 1, 319925, 7678200, 0.588],
    ['Lstm', 0.3, 'kme40_emb256_hid256', True, 1, 319925, 7678200, 0.5460932],
]

# 创建 DataFrame
columns = ['model_structure', 'train_ratio', 'model_name', 'hands_merged', 'frame_sampling', 'params_total', 'params_per_second', 'bleu']
df = pd.DataFrame(data, columns=columns)

# 新增一列用于分类颜色
def get_model_type(row):
    if '_multi-head' in row['model_name']:
        return 'Multi-head'
    elif row['hands_merged']:
        return 'Hand Merged'
    else:
        return 'Baseline'

df['model_type'] = df.apply(get_model_type, axis=1)

# 绘图
plt.figure(figsize=(12, 20))

scatter = plt.scatter(
    df['params_per_second'],
    df['bleu'],
    c=df['model_type'].astype('category').cat.codes,
    s=df['train_ratio'] * 200,  # 点的大小与训练集占比成正比
    alpha=0.7,
    cmap='viridis',
    edgecolor='k'
)

# 添加标签
for i, txt in enumerate(df['model_name']):
    plt.annotate(txt, (df['params_per_second'].iloc[i], df['bleu'].iloc[i]), fontsize=8)

# 设置标题和坐标轴
plt.title('Model Comparison: BLEU vs Parameter Throughput')
plt.xlabel('Parameters per Second')
plt.ylabel('BLEU Score')

handles, labels = scatter.legend_elements(prop="colors", num=len(df['model_type'].unique()))
# 使用真实的模型类型名称替换默认的标签
labels = ['Baseline', 'Hand Merged', 'Multi-head'] # 确保这里的顺序和分类结果一致
legend1 = plt.legend(handles, labels, title="Model Type", loc='lower right', bbox_to_anchor=(1.02, 0.55))

# 添加第二个图例：训练集占比
sizes = [0.2, 0.3, 0.5, 0.8] # 训练集比例列表
for size in sizes:
    plt.scatter([], [], c='gray', alpha=0.3, s=size*200, label=f'Train Ratio {size}')
legend2 = plt.legend(title='Train Ratio', loc='lower right', bbox_to_anchor=(1.02, 0.05))

# 注意：我们调用了两次 plt.legend()，但是只显示了第二次调用的结果。
# 因此，我们需要添加第一个图例，并确保它不会被第二个覆盖。
plt.gca().add_artist(legend1) # 添加第一个图例



plt.grid(True)
plt.tight_layout()
plt.show()
```
