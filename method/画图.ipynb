```python
import matplotlib.pyplot as plt

# 示例2D列表数据，包含两组(x, y)数据
data = [
    #第一条线，emb128，hid128
    [(40, 0.657), 
     (50, 0.658),
     (70, 0.655),
     (100, 0.641)],
    #第二条线，emb256,hid256
    [(20, 0.650), (40, 0.668), (50, 0.661),(70,0.6556),(100,0.655)],
    #第三条线，emb512,hid512
    [(30,0.647),(40,0.669),(70,0.664),(100,0.662)],
    #第四条线 emb 512 hid 256
    [(50,0.660),],
]

# 绘制每条折线
for d in data:
    x_values = [point[0] for point in d] # 提取x值
    y_values = [point[1] for point in d] # 提取y值
    plt.plot(x_values, y_values, marker='o') # 绘制折线，并在点上显示小圆圈标记

# 添加标题和轴标签
plt.title('2D')
plt.xlabel('emb')
plt.ylabel('bleu')
# plt.ylim(0.5, 1)

# 显示图表
plt.legend(['emb128 hid128', 'emb256 hid256','emb512 hid512','emb512 hid256']) # 如果有多条线，可以添加图例
plt.show()
```
