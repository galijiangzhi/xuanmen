```python
import os
import numpy as np
import matplotlib.pyplot as plt
```

```python
all_path = '../SLR_dataset/seq_txt'
```

```python
folder_list = os.listdir(all_path)
```

```python
folder_list = [os.path.join(all_path,i) for i in folder_list]
```

```python
def left_right_hand(data):
    left = data[0:63]
    right = data [63:]
    return left,right
```

```python
def list_proLR(data):
    #该函数用于将读取到的数据变成左右手格式
    result = []
    for i in data:
        left,right = left_right_hand(i)
        result.append([left,right])
    return result
```

```python
def file_statistics_data(data):
    #该函数输入分好左右手的数据，输出每个视频的有效率
    yes_data = 0
    no_data = 0
    for i in data:
        left = i[0]
        right = i[1]
        # print(left,right)
        if (len(set(left)) == 1) and (len(set(right)) == 1):
            no_data+=1
        else:
            yes_data+=1
            
    return yes_data/(yes_data+no_data)
```

```python
for folder in folder_list:
    data_list = [os.path.join(folder,filename) for filename in os.listdir(folder) if filename.endswith('.avi')]
    for file in data_list:
        # print(file)
        data = []
        with open(file,'r') as f:
            for line in f:
                float_values = [float(x) for x in line.split()]
                data.append(float_values)
        # print(len(data)) #二维数组 帧*126
        # print(data[0:10])
        data = list_proLR(data) #三维数组 帧*2*63
        statistice=file_statistics_data(data)
        if statistice < 0.74:
            os.remove(file)
            print(file,statistice)
```

```python
sum = 0
for folder in folder_list:
    data_list = [os.path.join(folder,filename) for filename in os.listdir(folder) if filename.endswith('.avi')]
    print(folder,len(data_list))
    sum+=len(data_list)

print(f'总数据量为{sum}')
```
